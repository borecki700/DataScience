Zadanie 26: System Monitoringu Temperatury (Data Cleaning)
Masz listę pomiarów z czujnika, ale dane są "brudne" (niektóre to błędy typu None lub teksty).

Python
pomiary = [22.5, 23.0, "ERROR", 24.5, None, 22.8, "MISSING", 25.0, 23.5]
Twoim zadaniem jest:

Stworzyć nową listę czyste_pomiary, która zawiera tylko liczby (float).

Jeśli napotkasz None lub tekst, pomiń go (użyj continue).

Obliczyć temperaturę minimalną, maksymalną oraz średnią (zaokrągloną do 1 miejsca po przecinku).

Wyświetlić raport:
Pomiary poprawne: [liczba]
Zakres: [min] - [max] st. C
Średnia: [średnia] st. C

```python
pomiary = [22.5, 23.0, "ERROR", 24.5, None, 22.8, "MISSING", 25.0, 23.5]
czyste_pomiary = []

for pom in pomiary:
    if pom is None:
        continue
    try:
        pom = float(pom)
        czyste_pomiary.append(pom)
    except (ValueError, TypeError):
        continue

min_temp = min(czyste_pomiary)
max_temp = max(czyste_pomiary)
srednia_temp = round(sum(czyste_pomiary) / len(czyste_pomiary), 1)

print(f"Pomiary poprawne: {len(czyste_pomiary)}")
print(f"Zakres: {min_temp} - {max_temp} st. C")
print(f"Średnia: {srednia_temp} st. C")
```

Obliczyć sumę jego koszyka.

Naliczyć rabat:

Jeśli suma > 1000 zł -> 10% rabatu.

Jeśli ktoś jest stałym klientem -> dodatkowe 5% rabatu (rabaty się sumują).

Wyświetlić dla każdego: [Imię]: Do zapłaty [kwota po rabacie] (Oszczędność: [kwota rabatu]).

```python
surowe_dane = [
    {"id": "101", "cena": "150.50", "status": "zrealizowano", "tagi": "elektronika;dom"},
    {"id": "102", "cena": "error", "status": "zrealizowano", "tagi": "ogrod"},
    {"id": "103", "cena": "300.00", "status": "anulowano", "tagi": "elektronika"},
    {"id": "104", "cena": "20.99", "status": "zrealizowano", "tagi": "biuro;dom;elektronika"},
    {"id": "105", "cena": "500.00", "status": "zrealizowano", "tagi": None}
]


raport_kategorii = {}

for dane in surowe_dane:
    if dane['status'] == "zrealizowano":
        try:
            cena = float(dane['cena'])
        except (ValueError, TypeError):
            continue
        
        if dane['tagi'] is not None:
            tagi_lista = dane['tagi'].split(";")
            for tag in tagi_lista:
                if tag not in raport_kategorii:
                    raport_kategorii[tag] = 0
                raport_kategorii[tag] += cena

print(raport_kategorii)
```

Zadanie 2: Algorytmika – Średnia Krocząca (Moving Average)
To klasyczne zadanie na kierunkach technicznych, używane do wygładzania wykresów giełdowych.

Dane:

Python
ceny_akcji = [10, 12, 15, 13, 16, 18, 14, 20, 22]
okno = 3
Polecenie:

Oblicz tzw. średnią kroczącą dla okna o rozmiarze 3.

Oznacza to, że musisz stworzyć listę srednie, gdzie:

Pierwszy element to średnia z [10, 12, 15]

Drugi element to średnia z [12, 15, 13]

Trzeci to średnia z [15, 13, 16] ... i tak dalej.

Musisz użyć pętli for oraz wycinania list (slicing). Pętla musi skończyć się tak, aby nie wyjść poza zakres listy.

Wyniki zaokrąglij do 2 miejsc po przecinku.

```python
ceny_akcji = [10, 12, 15, 13, 16, 18, 14, 20, 22]
okno = 3

srednie = []

for i in range(len(ceny_akcji)):
    print(ceny_akcji[i:i+okno])
    if i+okno-1 < 8:
        srednia_krocząca = round(float(sum(ceny_akcji[i:i+okno])/okno),2)
        srednie.append(srednia_krocząca)
    else:
        break

print(srednie)
    

    
    
```

Jeśli pasażer jest VIP -> dodatkowa zniżka 10% (liczona od kwoty, która została po poprzedniej zniżce).

Jeśli bagaż przekracza LIMIT_BAGAZU -> dopłata 50 zł za każdy kilogram nadmiaru.

Wyświetl raport: [Imię]: Cena końcowa: [X], Nadbagaż: [Tak/Nie].

Zadanie 4: Sterowanie Przepływem – Unikalny Scalacz (No Sets allowed!)
Na studiach często zabrania się używania gotowych funkcji (jak set()), aby sprawdzić, czy rozumiesz algorytm.

Dane:

Python
magazyn_A = ["laptop", "mysz", "monitor", "laptop", "klawiatura"]
magazyn_B = ["monitor", "drukarka", "mysz", "głośniki"]
Polecenie:

Stwórz listę wspolne_produkty – znajdź tylko te rzeczy, które są w OBU magazynach.

Stwórz listę wszystkie_unikalne – połącz obie listy tak, aby żaden produkt się nie powtarzał.

UWAGA: Nie możesz użyć set(). Musisz użyć pętli i operatora in.

```python
magazyn_A = ["laptop", "mysz", "monitor", "laptop", "klawiatura"]
magazyn_B = ["monitor", "drukarka", "mysz", "głośniki"]

wspolne_produkty = []
wszystkie_unikalne = []

for i in magazyn_A:
    if i in magazyn_B:
        wspolne_produkty.append(i)
    if i not in wszystkie_unikalne and i not in magazyn_B:
        wszystkie_unikalne.append(i)
for j in magazyn_B:
    if j not in magazyn_A:
        wszystkie_unikalne.append(j)

print(wspolne_produkty)
print(wszystkie_unikalne)
```
