Treść: Stwórz kalkulator BMI (Body Mass Index). Wzór: BMI = masa / (wzrost)^2

Zadanie 4: Manipulacja tekstem

Treść: Napisz program, który:

    Policzę liczbę słów

    Znajdzie najdłuższe słowo

    Zamieni wszystkie 'a' na '@'

```python
print("napisz słowa odzielone spacjami")
tekst = input()
print(f'Slowa w tekscie {tekst.split(" ")}')
print(tekst.replace("a","@"))

slowa = tekst.split(" ")
najdluzsze = ""

for slowo in slowa:
    if len(slowo) > len(najdluzsze):
        najdluzsze = slowo

print(f'Najdlusze slowo w zdaniu to {najdluzsze}')
```

Zadanie 8: Analiza liczby

Treść: Napisz program, który dla podanej liczby:

    Sprawdzi czy jest parzysta

    Obliczy jej kwadrat i pierwiastek

    Wyświetli jej ostatnią cyfrę

Wskazówki:

    Operator modulo % do sprawdzenia parzystości

    ** 0.5 do pierwiastka

    Konwersja na string i [-1] do ostatniej cyfry

```python
print('Podaj dowolną liczbe całkowitą')
liczba = int(input())

if liczba % 2 == 0:
    print('Liczba parzysta')
else:
    print('Liczba nieparzysta')

kwadrat = liczba**2
pierwiastek = float(liczba**0.5)

print(f'Kwadrat twojej liczby to {kwadrat}')
print(f'Pierwiastek Twojej liczby to {pierwiastek}')
```

Zadanie 9: Konwerter temperatury

Treść: Stwórz program konwertujący:

    Celsjusze na Fahrenheity: F = C * 9/5 + 32

    Fahrenheity na Celsjusze: C = (F - 32) * 5/9

Wskazówki:

    Zapytaj użytkownika o typ konwersji

    Użyj if/else do wyboru opcji

    Zaokrąglij wyniki

```python
print('Jaką konwersję chcesz uzyskać z F na C (napisz 1) czy z C na F (napisz 0)')

wybor = bool(input())
print(wybor)

if wybor:
    print('Wybrałeś F na C, Podaj stopnie F')
    stopnie = int(input())
    konwersja = round(float((stopnie - 32) *(5/9)),2)
    print(f'Tyle to stopni C {konwersja}')
else:
    print('Wybrałeś C na F, Podaj stopnie C')
    stopnie = int(input())
    konwersja = round(float((stopnie * (9/5)) + 32),2)
    print(f'Tyle to stopni F {konwersja}')
```
