Poziom 5: Wyzwanie praktyczne (Zadania z notatek)
Spróbuj połączyć wiedzę, by rozwiązać realne problemy opisane w Twoich materiałach.
Obsługa braków: Stwórz tablicę [10, 20, np.nan, 40, 50, np.nan]. 
Oblicz jej średnią tak, aby ignorowała wartości NaN. 
Następnie zastąp wszystkie NaN zerami.
Normalizacja: Napisz funkcję, która przyjmuje tablicę i "ściska" jej wartości do zakresu [0, 1] według wzoru z notatek: {data - min}/{max - min}

```python
tablica = np.array([10, 20, np.nan, 40, 50, np.nan])
srednia = np.nanmean(tablica)
print(srednia)
print(np.nan_to_num(tablica).mean())
tablica = np.nan_to_num(tablica)

def normalizacja(df):
    x = df.min()
    y = df.max()
    return (df-x)/(y-x)

normalizacja(tablica)


```
