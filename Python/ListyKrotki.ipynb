Zadanie 1: Głęboka modyfikacja (Lista w Słowniku)
Masz bazę studentów i ich przedmiotów.

Python
szkola = {
    "klasa_1A": {
        "uczniowie": [
            {"imie": "Jan", "oceny": [3, 4, 5]},
            {"imie": "Ania", "oceny": [5, 5, 6]}
        ]
    }
}
Polecenie: Napisz funkcję dodaj_ocene(baza, klasa, imie_ucznia, nowa_ocena).

Musisz "wejść" kolejno: Klucz klasy -> Klucz "uczniowie" -> Pętla po liście w poszukiwaniu imienia -> Klucz "oceny" -> .append().

```python
szkola = {
    "klasa_1A": {
        "uczniowie": [
            {"imie": "Jan", "oceny": [3, 4, 5]},
            {"imie": "Ania", "oceny": [5, 5, 6]}
        ]
    }
}


def dodaj_ocene(baza, klasa, imie, nowa_ocena):
    if klasa in baza.keys():
        for uczen in baza[klasa]['uczniowie']:
            if uczen["imie"] == imie:
                uczen["oceny"].append(nowa_ocena)
            return baza


dodaj_ocene(szkola,"klasa_1A",'Jan',5)
print(szkola)
```

Zadanie 2: Spłaszczanie danych (Matrix to List)
Analitycy często muszą "spłaszczyć" macierz (listę list) do jednej prostej listy.
Dane: macierz = [[1, 2, 3], [4, 5], [6, 7, 8, 9]]
Polecenie: Napisz funkcję flatten(matrix), która użyje zagnieżdżonych pętli for, aby zwrócić [1, 2, 3, 4, 5, 6, 7, 8, 9].

```python
macierz = [[1, 2, 3], [4, 5], [6, 7, 8, 9]]

def flatten(baza):
    pusta = []
    for i in baza:
        for j in i:
            pusta.append(j)
    return pusta

flatten(macierz)
```

Zadanie 3: Zaawansowane mapowanie (Analiza Logów)
Dostajesz listę krotek z danymi o temperaturze z różnych czujników.

Python
odczyty = [
    ("Salon", [20, 21, 20.5]),
    ("Kuchnia", [23, 22.5]),
    ("Sypialnia", [19, 20])
]
Polecenie: Napisz funkcję max_temp_raport(dane), która zwróci słownik, gdzie kluczem jest nazwa pomieszczenia, a wartością najwyższa temperatura z listy odczytów dla tego pomieszczenia.
Podpowiedź: Tu musisz połączyć pętlę po liście krotek z funkcją max() na liście wewnątrz krotki.

```python
odczyty = [
    ("Salon", [20, 21, 20.5]),
    ("Kuchnia", [23, 22.5]),
    ("Sypialnia", [19, 20])
]

def max_temp_raport(baza):
    slownik = {}
    for i in baza:
        slownik[i[0]]=max(i[1])
    return slownik

max_temp_raport(odczyty)
```

Zadanie 4: Filtracja zagnieżdżona
Z bazy produktów wybierz tylko te, których średnia cena w różnych sklepach przekracza 100 zł.

Python
produkty = [
    {"nazwa": "Klawiatura", "ceny_sklepy": [80, 90, 110]},
    {"nazwa": "Myszka", "ceny_sklepy": [150, 160]},
    {"nazwa": "Podkładka", "ceny_sklepy": [20, 30]}
]
Polecenie: Napisz funkcję filtruj_drogie(lista_prod), która zwróci listę nazw (stringów) produktów spełniających warunek.

```python
produkty = [
    {"nazwa": "Klawiatura", "ceny_sklepy": [80, 90, 110]},
    {"nazwa": "Myszka", "ceny_sklepy": [150, 160]},
    {"nazwa": "Podkładka", "ceny_sklepy": [20, 30]}
]

def filtruj_drogie(lista_prod):
    pusta = []
    for i in lista_prod:
        if sum(i['ceny_sklepy'])/len(i['ceny_sklepy']) >100:
            pusta.append(i['nazwa'])
    return pusta

filtruj_drogie(produkty)
```

Zadanie: System Analizy Sprzedaży (The Grand Master)
Dostajesz listę transakcji z błędami i brakami. Twoim zadaniem jest stworzyć "silnik", który wypluje gotowy raport.

Dane wejściowe:

Python
transakcje = [
    {"id": 1, "produkt": "smartfon", "ceny": [2000, 2100, "brak"], "region": "Północ"},
    {"id": 2, "produkt": "laptop", "ceny": [3500, 3600, 3400], "region": "Południe"},
    {"id": 3, "produkt": "smartfon", "ceny": [1900, "błąd", 2050], "region": "Północ"},
    {"id": 4, "produkt": "monitor", "ceny": [800, 850], "region": "Południe"},
]
Twoja funkcja procesuj_sprzedaz(dane, filtr_funkcja) ma:

Wyczyścić dane: Dla każdego produktu stwórz listę cen, ignorując stringi (takie jak "brak" czy "błąd"). Zostaw tylko liczby.

Obliczyć średnią: Dla każdego produktu oblicz średnią z jego dostępnych cen.

Zastosować filtr: Użyj parametru filtr_funkcja (przekazanego jako lambda), aby wybrać tylko te produkty, których średnia cena spełnia warunek.

Zasugerować wynik: Funkcja ma zwrócić słownik pogrupowany według regionów, gdzie wartością jest lista nazw produktów, które przeszły filtr.

Przetestuj funkcję tak:
Wyciągnij produkty, których średnia cena jest wyższa niż 1000 zł.

Oczekiwany wynik dla tych danych:
{'Północ': ['smartfon', 'smartfon'], 'Południe': ['laptop']}

```python
transakcje = [
    {"id": 1, "produkt": "smartfon", "ceny": [2000, 2100, "brak"], "region": "Północ"},
    {"id": 2, "produkt": "laptop", "ceny": [3500, 3600, 3400], "region": "Południe"},
    {"id": 3, "produkt": "smartfon", "ceny": [1900, "błąd", 2050], "region": "Północ"},
    {"id": 4, "produkt": "monitor", "ceny": [800, 850], "region": "Południe"},
]

def procesuj_sprzedaz(dane,filtr_funkcja):
    #czyszczenie danych
    slownik = {}
    for i in dane:
        for j in i['ceny']:
            try:
                int(j)
            except ValueError:
                i['ceny'].remove(j)
                continue
        i['średnia'] = float(sum(i['ceny'])/len(i['ceny']))
    #zastosowanie lambda
    for i in dane:
        if filtr_funkcja(i['średnia']):
            slownik.setdefault(i["region"], []).append(i["produkt"])

    return slownik

    
procesuj_sprzedaz(transakcje, lambda x : x>1000)
```
