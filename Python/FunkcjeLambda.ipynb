Zadanie 3: Wykrywanie Wartości Odstających (Outliers)
W statystyce często odrzucamy dane, które za bardzo odbiegają od średniej.
Polecenie: Napisz funkcję znajdz_odstajace(pomiary, prog=2).

Funkcja oblicza średnią z listy pomiary.

Funkcja znajduje odchylenie standardowe (możesz użyć m.sqrt z modułu math oraz wzoru: sqrt(suma((x - srednia)**2) / ilosc)).

Funkcja ma zwracać listę wartości, które są większe niż srednia + prog * odchylenie lub mniejsze niż srednia - prog * odchylenie.

Przetestuj na: [10, 12, 11, 13, 100, 11, 12, 9, 105, 10].

```python
import math as m

def znajdz_odstajace(pomiary, prog=2):
    srednia = sum(pomiary)/len(pomiary)
    
    s = 0
    for x in pomiary:
        s += (x - srednia) ** 2
        
    odchylenie = m.sqrt(s/len(pomiary))
    
    lista = []
    for i in pomiary:
        if i > srednia + prog * odchylenie or i < srednia - prog * odchylenie:
            lista.append(i)
    
    return lista


dane = [10, 12, 11, 13, 100, 11, 12, 9, 105, 10]
print(znajdz_odstajace(dane))
```

Zadanie 1: Analiza szeregów czasowych (Moving Average Pro)
W analizie danych giełdowych rzadko liczymy jedną średnią. Często potrzebujemy funkcji, która "wyczyści" dane z szumu.

Polecenie: Napisz funkcję smooth_signal(data, window_size=3).

Funkcja ma przyjmować listę liczb i rozmiar okna.

Dla każdego elementu (od którego da się wyznaczyć pełne okno), oblicz średnią z window_size elementów, ale uwaga:

Jeśli w oknie trafi się wartość None (brak danych), funkcja ma ją pominąć i policzyć średnią z pozostałych elementów w tym oknie.

Jeśli okno zawiera same wartości None, wstaw do wynikowej listy 0.

Przetestuj na: [10, None, 12, 15, 18, None, 20], okno = 3.

```python
def smooth_signal(data, window_size=3):
    wynik = []
    
    for i in range(len(data) - window_size + 1):
        okno = data[i:i+window_size]
        
        if not okno:
            break
        
        ile_none = okno.count(None)
        
        suma = sum(x for x in okno if x is not None)
        
        if window_size - ile_none > 0:
            srednia = suma / (window_size - ile_none)
        else:
            srednia = 0
            
        wynik.append(srednia)
    
    return wynik

lista = [10, None, 12, 15, 18, None, 20]
okno = 3

smooth_signal(lista,okno)
```

Funkcja ma zwrócić zagnieżdżony słownik: {osoba: {miesiac: suma_kwot}}.

Przykład wyniku: {"Ania": {"Styczeń": 150, "Luty": 150}, "Tomek": {"Styczeń": 200}}.

Wykorzystaj pętle i instrukcje warunkowe wewnątrz funkcji.

```python
logi = [
    {"osoba": "Ania", "miesiac": "Styczeń", "kwota": 100},
    {"osoba": "Tomek", "miesiac": "Styczeń", "kwota": 200},
    {"osoba": "Ania", "miesiac": "Luty", "kwota": 150},
    {"osoba": "Ania", "miesiac": "Styczeń", "kwota": 50},
]

def create_pivot(data_list):
    wynik = {}
    
    for wpis in data_list:
        osoba = wpis['osoba']
        miesiac = wpis['miesiac']
        kwota = wpis['kwota']
        
        # KROK 1: Sprawdź czy osoba jest w słowniku, jeśli nie - dodaj ją z pustym słownikiem
        if osoba not in wynik:
            wynik[osoba] = {}
            
        # KROK 2: Sprawdź czy miesiąc jest u tej osoby, jeśli nie - zainicjuj 0
        if miesiac not in wynik[osoba]:
            wynik[osoba][miesiac] = 0
            
        # KROK 3: Dodaj kwotę do konkretnego "pudełka"
        wynik[osoba][miesiac] += kwota
        
    return wynik
        
    
create_pivot(logi)
```

Zadanie 4: Walidator i Normalizator (Higher-Order Function)
W Data Science często budujemy "pipelines". Napisz funkcję, która aplikuje zestaw transformacji na liście danych.

Polecenie: Napisz funkcję preprocess_pipeline(data, transformations).

data to lista stringów reprezentujących liczby (np. [" 1.5 ", "2.0", " błąd ", " 3.7"]).

transformations to lista funkcji (lub lambd), które mają być wykonane po kolei.

Twoja funkcja powinna:

Spróbować zamienić każdy element na float. Jeśli się nie da (ValueError) – pominąć element.

Dla każdej poprawnej liczby wykonać po kolei wszystkie funkcje z listy transformations.

Przetestuj: Podaj listę [" 10 ", " 20 ", "30"] i transformacje: [lambda x: x*2, lambda x: x+5]. Wynik powinien wynosić [25.0, 45.0, 65.0].

```python
def preprocess_pipeline(data, transformations):
    dane_czyste = []
    for i in data:
        i = i.strip()
        try:
            float(i)
            dane_czyste.append(float(i))
        except ValueError:
            continue
    
    for i in transformations:
        dane_czyste = list(map(i, dane_czyste))
    
    return dane_czyste


lista = [" 10 ", " 20 ", "30"]
transformacje = [lambda x: x*2, lambda x: x+5]
preprocess_pipeline(lista, transformacje)
```

```python
szkola = {
    "klasa_1A": {
        "uczniowie": [
            {"imie": "Jan", "oceny": [3, 4, 5]},
            {"imie": "Ania", "oceny": [5, 5, 6]}
        ]
    }
}

def dodaj_ocene(baza, klasa, imie_ucznia, nowa_ocena):
    # sprawdź czy klasa istnieje
    if klasa not in baza:
        return baza

    uczniowie = baza[klasa].get("uczniowie", [])
    for uczen in uczniowie:
        if uczen.get("imie") == imie_ucznia:
            if "oceny" not in uczen or not isinstance(uczen["oceny"], list):
                uczen["oceny"] = []
            uczen["oceny"].append(nowa_ocena)
            break  # zakładamy, że imię występuje tylko raz
        
    return baza

dodaj_ocene(szkola,"klasa_1A",'Jan',5)
```
