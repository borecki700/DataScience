Zadanie 14: Kalkulator BMI (Math & Conversion)Cel:

Oblicz BMI według wzoru: 
$BMI = \frac{masa}{wzrost^2}$.

Zadanie 20: Dynamiczny środek napisu
Napisz kod, który przyjmuje dowolny string i wypisuje jego środkowy znak.
Dla ułatwienia: zakładamy, że string ma nieparzystą liczbę znaków (np. "Python" ma 6, weź "Pythons" - 7). Użyj len(s) // 2 do znalezienia środkowego indeksu.

```python
tekst = input("Podaj tekst o dl nieparzystej: ")
if len(tekst) % 2 == 1:
    print("Długość tekstu jest nieparzysta")
    srodek = len(tekst) // 2 
    print(srodek)
    print(f"Środek tekstu: {tekst[srodek]}")
else:
    print("Długość tekstu jest parzysta")
```
