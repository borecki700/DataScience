Zadanie 4: Wycinanie (Slicing)

Teoria: Działa identycznie jak w stringach: lista[start:stop].

Polecenie:
Mamy listę pracowników: kadra = ["Szef", "Manager", "Analityk", "Programista", "Stażysta"].
Wypisz:

Pierwsze dwie osoby ("Zarząd").

Ostatnie dwie osoby (użyj ujemnych indeksów).

Co drugą osobę z listy.

```python
kadra = ["Szef", "Manager", "Analityk", "Programista", "Stażysta"]
zarząd = kadra[0:2]
print(zarząd)
dwie_osoby = kadra[-2:]
print(dwie_osoby)
co_druga_osoba = kadra[::2]
print(co_druga_osoba)
```

Zadanie 5: Usuwanie (Pop vs Remove)
Teoria:


.pop(index) usuwa element po indeksie i go zwraca (można go zapisać do zmiennej).


.remove(wartość) usuwa pierwsze wystąpienie konkretnej wartości (nic nie zwraca).

Polecenie:
Masz listę: do_zrobienia = ["pranie", "nauka pythona", "zakupy", "spanie", "nauka pythona"].

Usuń ostatni element za pomocą .pop() i przypisz go do zmiennej zrobione. Wyświetl: "Właśnie zrobiłem: [zrobione]".

Usuń z listy zadanie "nauka pythona" (pierwsze wystąpienie) za pomocą .remove().

Wyświetl, co zostało na liście.

```python
do_zrobienia = ["pranie", "nauka pythona", "zakupy", "spanie", "nauka pythona"]
zrobione = do_zrobienia.pop()
print(f"Wlaśnie zrobiłem: {zrobione}")
do_zrobienia.remove("nauka pythona")
print(do_zrobienia)
```

Zadanie 8: Rozpakowywanie krotki (Unpacking)

Zadanie 10: Pułapka kopiowania (Copy vs Reference)
Teoria: a = b tworzy tylko "skrót" do tej samej listy. Jeśli zmienisz a, zmieni się też b! Aby tego uniknąć, używamy .copy() .
Polecenie:

Stwórz listę a = [1, 2, 3].

Przypisz ją do zmiennej b znakiem równości: b = a.

Przypisz ją do zmiennej c używając metody kopiowania: c = a.copy().

Zmień pierwszy element listy a na 99 (a[0] = 99).

Wyświetl listę b i listę c.

W komentarzu napisz, dlaczego b się zmieniło, a c pozostało [1, 2, 3].

```python
a = [1, 2, 3]
b = a 
c = a.copy()
a[0] = 99
print(b)
print(c)
#b się zmieniło gdyż jest tym samym obiektem co a, natomiast c jest kopią i dlatego się nie zmieniło
```

Zadanie 12: Metody słownika (Keys, Values, Items)
Teoria:

.keys() – zwraca listę kluczy.

.values() – zwraca listę wartości.

.items() – zwraca pary (klucz, wartość) jako krotki.
Polecenie:
Masz słownik: oceny = {"matematyka": 4, "fizyka": 5, "chemia": 3}.

Wypisz tylko nazwy przedmiotów (klucze).

Wypisz sumę ocen (użyj sum() na wartościach słownika).

```python
oceny = {"matematyka": 4, "fizyka": 5, "chemia": 3}
print(list(oceny.keys()))
print(list(oceny.values()))
print(list(oceny.items()))
print(f"Suma ocen: {sum(oceny.values())}")
```

Zadanie 15: Usuwanie ze słownika (del vs pop)
Teoria: del slownik["klucz"] usuwa bezpowrotnie. slownik.pop("klucz") usuwa, ale pozwala zapisać wartość do zmiennej.
Polecenie:
Masz słownik: portfolio = {"BTC": 50000, "ETH": 3000, "LTC": 100}.

Usuń "LTC" za pomocą del.

Usuń "ETH" za pomocą .pop() i zapisz jego wartość do zmiennej cena_eth.

Wyświetl słownik oraz zmienną cena_eth.

```python
portfolio = {"BTC": 50000, "ETH": 3000, "LTC": 100}
del portfolio["LTC"]
print(portfolio)
cena_eth = portfolio.pop("ETH")
print(portfolio)
print(cena_eth)
```

ceny_str = ["12.50", "150.00", "29.99", "4.50"]

Masz listę identyfikatorów sesji, w której niektóre się powtarzają:
sesje = [101, 102, 101, 105, 102, 108, 101]

Twoim zadaniem jest:

Stworzyć listę unikalne_sesje, w której każdy identyfikator występuje tylko raz (użyj tego, co wiesz o usuwaniu elementów lub zamianie typów – podpowiedź: w Twoim PDF-ie nie było jeszcze zbiorów/sets, więc spróbuj to przemyśleć manualnie lub sprytnie wykorzystać dostępne metody).

firma = {
    "nazwa": "TechCorp",
